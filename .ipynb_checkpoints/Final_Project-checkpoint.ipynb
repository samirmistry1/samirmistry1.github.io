```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches

##creating a new dataframe to hold the values
emptyData = np.empty((390,3))
emptyData[:] = np.nan
valData = pd.DataFrame(emptyData)

## labeling the new dataframe columns
valData.columns = ['height', 'winColor','age'] 

## iterating through the original dataframe and inserting values into new one based on winner
valIndex = 0
for index,row in firstData.iterrows():
    winner = firstData.get_value(index, 'winner')
    redHeight = firstData.get_value(index, 'R_Height')
    blueHeight = firstData.get_value(index, 'B_Height')
    redAge = firstData.get_value(index, 'R_Age')
    blueAge = firstData.get_value(index, 'B_Age')
    tempRAge = str(redAge)
    tempBAge = str(blueAge)
    if(winner == 'red'):
        if(tempRAge != '0.0'):
            valData.loc[valIndex,'height'] = redHeight
            valData.loc[valIndex,'winColor'] = 'g'
            valData.loc[valIndex,'age'] = redAge
            valIndex = valIndex + 1
        if(tempBAge != '0.0'):
            valData.loc[valIndex,'height'] = blueHeight
            valData.loc[valIndex,'winColor'] = 'r'
            valData.loc[valIndex,'age'] = blueAge
            valIndex = valIndex + 1
    else:
        if(tempRAge != '0.0'):
            valData.loc[valIndex,'height'] = blueHeight
            valData.loc[valIndex,'winColor'] = 'g'
            valData.loc[valIndex,'age'] = blueAge
            valIndex = valIndex + 1
        if(tempBAge != '0.0'):
            valData.loc[valIndex,'height'] = redHeight
            valData.loc[valIndex,'winColor'] = 'r'
            valData.loc[valIndex,'age'] = redAge
            valIndex = valIndex + 1

## plotting the new dataframe
plt.scatter([x for x in valData.age], valData.height, c = valData.winColor)
## adding a title and labels to the plot
plt.title('Age vs. Height')
winner = mpatches.Patch(color='green', label='Winner')
loser = mpatches.Patch(color='red', label='Loser')
plt.legend(handles=[winner,loser])
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

Intuitivley, we wanted to see whether or not the amount of strikes landed vs. attempts would be a good predictor on who would win the fight. Although the data does not show the green(winners) in a different area than the reds(losers), it does show a strong correlation between strike attempts and strikes landed. 

```python
##creating a temporary dataframe to hold new values
strikesLandedDf = firstData

strikesLandedDf.fillna(value=0,inplace=True)

##adding the landed strike attempts in one column
strikesLandedDf['BLandsAdded'] = strikesLandedDf['B__Round1_Strikes_Distance Head Strikes_Landed']+strikesLandedDf['B__Round2_Strikes_Distance Head Strikes_Landed'] +strikesLandedDf['B__Round3_Strikes_Distance Head Strikes_Landed']
strikesLandedDf['RLandsAdded'] = strikesLandedDf['R__Round1_Strikes_Distance Head Strikes_Landed']+strikesLandedDf['R__Round2_Strikes_Distance Head Strikes_Landed'] +strikesLandedDf['R__Round3_Strikes_Distance Head Strikes_Landed']

##adding the attempted strike attempts in another column
strikesLandedDf['BAttemptAdded'] = strikesLandedDf['B__Round1_Strikes_Distance Head Strikes_Attempts']+strikesLandedDf['B__Round2_Strikes_Distance Head Strikes_Attempts'] +strikesLandedDf['B__Round3_Strikes_Distance Head Strikes_Attempts']
strikesLandedDf['RAttemptAdded'] = strikesLandedDf['R__Round1_Strikes_Distance Head Strikes_Attempts']+strikesLandedDf['R__Round2_Strikes_Distance Head Strikes_Attempts'] +strikesLandedDf['R__Round3_Strikes_Distance Head Strikes_Attempts']

##creating a new empty dataframe to hold data
emptyData = np.empty((390,3))
emptyData[:] = np.nan
valData2 = pd.DataFrame(emptyData)

##labeling the columns of the dataframe
valData.columns = ['lands', 'attempts','color']

##iterating through the original dataframe to get values based on the winner
valIndex2 = 0
for index, row in strikesLandedDf.iterrows():
    winner = strikesLandedDf.get_value(index, 'winner')
    BLand = strikesLandedDf.get_value(index, 'BLandsAdded')
    RLand = strikesLandedDf.get_value(index, 'RLandsAdded')
    BAttempt = strikesLandedDf.get_value(index, 'BAttemptAdded')
    RAttempt = strikesLandedDf.get_value(index, 'RAttemptAdded')
    if(winner == 'red'):
        valData2.loc[valIndex2,'lands'] = RLand
        valData2.loc[valIndex2,'attempts'] = RAttempt
        valData2.loc[valIndex2,'color'] = 'g'
        valIndex2 = valIndex2 + 1
        valData2.loc[valIndex2,'lands'] = BLand
        valData2.loc[valIndex2,'attempts'] = BAttempt
        valData2.loc[valIndex2,'color'] = 'r'
        valIndex2 = valIndex2 + 1
    else:
        valData2.loc[valIndex2,'lands'] = BLand
        valData2.loc[valIndex2,'attempts'] = BAttempt
        valData2.loc[valIndex2,'color'] = 'g'
        valIndex2 = valIndex2 + 1
        valData2.loc[valIndex2,'lands'] = RLand
        valData2.loc[valIndex2,'attempts'] = RAttempt
        valData2.loc[valIndex2,'color'] = 'r'
        valIndex2 = valIndex2 + 1
    
## plotting the new data
plt.scatter([x for x in valData2.attempts], valData2.lands, c = valData2.color)
##adding labels to the data
plt.title('Attempts vs. Lands')
winner = mpatches.Patch(color='green', label='Winner')
loser = mpatches.Patch(color='red', label='Loser')
plt.legend(handles=[winner,loser])
plt.xlabel('Attempts')
plt.ylabel('Lands')
plt.show()
```

```python
##removing the unused round 4 and round 5 data

data2 = data[np.isfinite(data['B__Round4_Grappling_Reversals_Landed'])]

data['remove'] = 'no'

for index,row in data2.iterrows():
    data.loc[index,'remove'] = 'yes'

data = data[data.remove == 'no']


data3 = data[np.isfinite(data['R__Round4_Grappling_Reversals_Landed'])]

data['remove2'] = 'no'

for index,row in data3.iterrows():
    data.loc[index,'remove2'] = 'yes'

data = data[data.remove2 == 'no']

##changing the NaNs to 0

data.fillna(value=0,inplace=True)

##computing the random forest code

##removing columns with strings in it
X_items = data.drop(['winner','remove','remove2','winby', 'B_ID', 'R_ID', 'B_Name', 'R_Name', 'Date',
                     'BStreak', 'B_HomeTown', 'R_HomeTown', 'B_Location', 'R_Location'], axis = 1)
Y_items = data['winner']

X_columns, testX, Y_columns, testY = train_test_split(X_items, Y_items, test_size=0.2, random_state=0)
randomfor = RandomForestClassifier(max_features="sqrt",n_estimators = 700,max_depth = None, n_jobs=-1)
randomfor.fit(X_columns,Y_columns)
prediction = randomfor.predict(testX)

random2 = accuracy_score(testY,prediction)

print('Random Forest with removal accuracy: ' + str(random2))
```
